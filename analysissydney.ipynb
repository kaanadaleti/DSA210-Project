```python
import pandas as pd

df = pd.read_csv("/Users/kaanadaleti/Desktop/dsa proje/SydneyAmerican.csv")
df.head()
```

```python
df["Time"] = pd.to_datetime(df["Time"])
df = df.sort_values("Time")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(df["Time"], df["American Eagle"], label="American Eagle")
plt.plot(df["Time"], df["Sydney Sweeney"], label="Sydney Sweeney")

plt.legend()
plt.xticks(rotation=45)
plt.title("Search Interest Over Time")
plt.show()
```

```python
event_start = "2025-07-27"
event_end = "2025-08-03"

event_df = df[(df["Time"] >= event_start) & (df["Time"] <= event_end)]
event_df
```

```python
before = df[df["Time"] < event_start]["American Eagle"]
after = df[df["Time"] >= event_start]["American Eagle"]

print(before.mean(), after.mean())
```

```python
df[["American Eagle", "Sydney Sweeney"]].corr()
```

A very strong positive correlation (r ≈ 0.93) is observed between Sydney Sweeney and American Eagle search interest.

This suggests a synchronized movement in public attention, where increases in celebrity visibility are closely mirrored by increases in brand interest. Unlike cases where attention shifts from brand to individual, this indicates a co-movement effect.

```python
plt.scatter(df["Sydney Sweeney"], df["American Eagle"])
plt.title("Correlation: Sydney Sweeney vs American Eagle")
plt.show()
```

Although the scatter plot is not perfectly linear due to the small sample size and volatility in Google Trends data, the strong positive correlation (r ≈ 0.93) indicates a clear co-movement between Sydney Sweeney and American Eagle search interest.
