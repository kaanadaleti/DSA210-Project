```python
from scipy.stats import ttest_ind

# tarih sütununu datetime yap
df['Time'] = pd.to_datetime(df['Time'])

# event date
event_date = '2022-10-25'

before = df[df['Time'] < event_date]['ADIDAS']
after = df[df['Time'] >= event_date]['ADIDAS']

t_stat, p_value = ttest_ind(before, after)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

## Correlation Analysis

Correlation analysis reveals that search interest for Kanye West and Yeezy is moderately positively correlated (r ≈ 0.56).

In contrast, the correlation between Kanye West and Adidas is very weak, suggesting that attention does not strongly transfer to the brand.
